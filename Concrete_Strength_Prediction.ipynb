```python
for feature in data.columns:
    if feature =="Age (day)":
        pass
    else :
        data2=data.copy()
        bar=sbn.distplot(data2[feature])
        bar.legend(["Skewness: {:.2f}".format(data2[feature].skew())])
        plt.xlabel(feature)
        plt.ylabel("Count")
        plt.title(feature)
        plt.show()
        
```

```python
for feature in data.columns:
    data2=data.copy()
    if 0 in data2[feature].unique():
        pass
    else:
        data2[feature]=np.log(data2[feature])
        data2.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
for feature in data.columns:
    if feature == "Water (component 4)(kg in a m^3 mixture)":
        pass
    if feature== "Age (day)":
        pass
    elif feature == "Coarse Aggregate (component 6)(kg in a m^3 mixture)":
        pass
    elif feature == "Fine Aggregate (component 7)(kg in a m^3 mixture)":
        pass
    else:
        data[feature]=np.log1p(data[feature])
```

```python
q_1=data.quantile(0.25)
q_3=data.quantile(0.75)
IQR=q_3-q_1
data=data[~((data<(q_1-1.5*IQR)) | (data>(q_3+1.5*IQR))).any(axis=1)]
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=5)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
reg=LinearRegression()
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
score=r2_score(y_test,y_pred)
print(f"value of R^2 is = {score}")
adjusted_score=(1-(1-score)*(len(y_train)-1)/(len(y_train)-len(data.columns)-1))
print(f"value of adjusted R^2 is = {adjusted_score}")
```
